```python
# importing important libraries for web scraping

from urllib.request import urlopen
from bs4 import BeautifulSoup as soup
import re
```

```python
## reading the url
```

```python
searchstring = 'laptop_under_50000'
url = 'https://www.flipkart.com/search?q='
```

```python
flipkart_url = url+searchstring
flipkart_url
```

```python
httpObject = urlopen(flipkart_url)
webdata = httpObject.read()# this data in a byte formate
type(webdata)
```

```python
# makeing a soup object 
soupdata = soup(webdata)
type(soupdata)
```

```python
# let's check the soup directory
#dir(soup)
```

```python
# all html data inside this soupdata
```

```python
containers = soupdata.findAll('div',{'class':'_2kHMtA'})
```

```python
print(f' the length of the containers are {len(containers)}, and the types of the containers is :{type(containers)}')

# ResultSet mean like list type
```

# 2. Let's Grab The Product Stars 

```python
for container in containers:
    stars = container.find('div',{'class':'_3LWZlK'})
    try:
        print(stars.text)
    except:
        print(stars)
    print('')
```

# 3. Now grabing the rating and reviews of the products


```python
for container in containers:
    Rating_Reviews = container.find('span',{'class':'_2_R_DZ'})
    try:
        print(Rating_Reviews.text)
        
    except:
        print(Rating_Reviews)
```

### Let's Seprate the ratings and reviews

```python
# Rating
for container in containers:
    Rating_Reviews = container.find('span',{'class':'_2_R_DZ'})
    try:
        Rating = Rating_Reviews.text.split('&')[0].split(" ")[0]
        print(Rating.replace(",",""))
    except:
        print('0 Rating')
```

```python
# Rating
for container in containers:
    Rating_Reviews = container.find('span',{'class':'_2_R_DZ'})
    try:
        Reviews = Rating_Reviews.text.split('&')[1]
        print(Reviews.split(" ")[0])
    except:
        print(' 0 Reviews')
```

# Grabing the Product Information

```python
for container in containers:
    product_info = container.findAll('li',{'class':'rgWa7D'})
    Processor = product_info[0].text
    RAM = product_info[1].text
    OS = product_info[2].text
    Storage = product_info[3].text
    Warranty = product_info[5].text[0]
    print("",Processor,'\n',RAM,'\n',OS,'\n',Storage,'\n',Warranty)
    print()
```

```python
for container in containers:
    ImageURL = container.img
    print(ImageURL.get('src'))
    print('\n')
```

#  Final Opration
### Here i am combining all operation and create a csv file  :)

```python
f = open('laptop_info1.csv','wb')
f.write("product_name,Stars,Ratings,Reviews,Current_price,Actual_Price,Processor,RAM,OS,Storage,Warranty,ImageURL\n".encode())
i = 0
for container in containers:
    i=i+1
    print("Product No.-",i)
    product = container.findAll('div',{'class':'_4rR01T'})
    print(product[0].text.split("-")[0])
    product_name =product[0].text.split("-")[0]
    
    # finding stars
    star = container.find('div',{'class':'_3LWZlK'})
    try:
        Stars = star.text
    except:
        Stars = 0
    
    # finding out the rating and reviews
    Rating_Reviews = container.find('span',{'class':'_2_R_DZ'})
    try:
        ratings = Rating_Reviews.text.split("&")[0].split(' ')[0].replace(',','').strip()
        reviews = Rating_Reviews.text.split("&")[1].split(' ')[0].replace(',','').strip()
        print(reviews,ratings)
    except:
        reviews =0
        ratings =0
        print('0','0')
    #current price
    current_price = container.find('div',{'class':'_30jeq3 _1_WHN1'})
    Price = current_price.text.replace(',','').replace('','').strip()
    print(Price)
    
    #Actual price
    Actual_Price = container.find('div',{'class':'_3I9_wc _27UcVY'})
    try:
        MRP = Actual_Price.text.replace(',','').replace('','').strip()
        print(MRP)
    except:
        MRP=0
        print('0')
    
    #product informatio
    product_info = container.findAll('li',{'class':'rgWa7D'})
    Processor = product_info[0].text.strip()
    RAM = product_info[1].text.strip()
    OS = product_info[2].text.strip()
    Storage = product_info[3].text.strip()
    Warranty = product_info[5].text.split(',')[0].strip()
    print("",Processor,'\n',RAM,'\n',OS,'\n',Storage,'\n',Warranty)

    #image url
    URL = container.img
    ImageURL=URL.get('src')
    print(ImageURL)
    print('\n')
    
    f.write(f"{product_name},{Stars},{ratings},{reviews},{Price},{MRP},{Processor},{RAM},{OS},{Storage},{Warranty},{ImageURL}\n".encode())
f.close()
```
